```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import time

filepath = '/workspace/examples/digital_twins/drobotti_movement_simulation/target/outputs.csv'
csvdata = json.loads(pd.read_csv(filepath).to_json())

x = []
y = []
for i in range(0,len(csvdata['{kin}.kinematicModel.x'])):
    x.append(csvdata['{kin}.kinematicModel.x'][str(i)])
    y.append(csvdata['{kin}.kinematicModel.y'][str(i)])
plt.figure(figsize=(15,15))
plt.plot(x, y, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X,Y position')
plt.show()

    
```
